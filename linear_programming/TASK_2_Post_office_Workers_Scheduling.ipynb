# TASK 2: Post office Workers Scheduling

```python
# Model definition
model = pyomo.ConcreteModel();

# Variables definition
model.x1 = pyomo.Var();
model.x2 = pyomo.Var();
model.x3 = pyomo.Var();
model.x4 = pyomo.Var();
model.x5 = pyomo.Var();
model.x6 = pyomo.Var();
model.x7 = pyomo.Var();

# Objective function definition
model.obj = pyomo.Objective(expr = model.x1 + model.x2 + model.x3 + model.x4
                            + model.x5 + model.x6 + model.x7, sense = pyomo.minimize);

# Constraint definition
def constraint1(model):
    return model.x1 + model.x4 + model.x5 + model.x6 + model.x7 >= 17
model.eq1 = pyomo.Constraint(rule = constraint1, doc='Monday requirement');

def constraint2(model):
    return model.x1 + model.x2 + model.x5 + model.x6 + model.x7 >= 13
model.eq2 = pyomo.Constraint(rule = constraint2, doc='Tuesday requirement');

def constraint3(model):
    return model.x1 + model.x2 + model.x3 + model.x6 + model.x7 >= 15
model.eq3 = pyomo.Constraint(rule = constraint3, doc='Wednesday requirement');

def constraint4(model):
    return model.x1 + model.x2 + model.x3 + model.x4 + model.x7 >= 19
model.eq4 = pyomo.Constraint(rule = constraint4, doc='Thursday requirement');

def constraint5(model):
    return model.x1 + model.x2 + model.x3 + model.x4 + model.x5 >= 14
model.eq5 = pyomo.Constraint(rule = constraint5, doc='Friday requirement');

def constraint6(model):
    return model.x2 + model.x3 + model.x4 + model.x5 + model.x6 >= 16
model.eq6 = pyomo.Constraint(rule = constraint6, doc='Saturday requirement');

def constraint7(model):
    return model.x3 + model.x4 + model.x5 + model.x6 + model.x7 >= 11
model.eq7 = pyomo.Constraint(rule = constraint7, doc='Sunday requirement');

def constraint8(model):
    return model.x1 >= 0
model.eq8 = pyomo.Constraint(rule = constraint8, doc='Non neg for monday');

def constraint9(model):
    return model.x2 >= 0
model.eq9 = pyomo.Constraint(rule= constraint9, doc = 'Non neg for tuesday');

def constraint10(model):
    return model.x3 >= 0
model.eq10 = pyomo.Constraint(rule= constraint10, doc = 'Non neg for wednesday');

def constraint11(model):
    return model.x4 >= 0
model.eq11 = pyomo.Constraint(rule = constraint11, doc='Non neg for thursday');

def constraint12(model):
    return model.x5 >= 0
model.eq12 = pyomo.Constraint(rule= constraint12, doc = 'Non neg for friday');

def constraint13(model):
    return model.x6 >= 0
model.eq13 = pyomo.Constraint(rule= constraint13, doc = 'Non neg for saturday');

def constraint14(model):
    return model.x7 >= 0
model.eq14 = pyomo.Constraint(rule= constraint14, doc = 'Non neg for sunday');


# Solver options
results = pyomo.SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model);

# Printing results
results.write();
print('\n RESULTS: \n');
print('Optimal number of workers = ', model.obj());

print('Number of workers assigned to Monday: ', model.x1());
print('Number of workers assigned to Tuesday: ', model.x2());
print('Number of workers assigned to Wednesday: ', model.x3());
print('Number of workers assigned to Thursday: ', model.x4());
print('Number of workers assigned to Friday: ', model.x5());
print('Number of workers assigned to Saturday: ', model.x6());
print('Number of workers assigned to Sunday: ', model.x7());

```
